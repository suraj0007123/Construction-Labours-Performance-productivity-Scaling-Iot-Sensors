# Dropping the Columns Which are Un-necessary

```python
data.drop(['Employee ID','Name','Time In','Time Out'],axis=1 ,inplace=True)
```

```python
data.dtypes
```

```python
features_numerical = ['Age','Experience','Heart Beat','Temperature','Working Hours','Remaining Working Hours','Beats Per Minute','Body Temperature']
```

```python
for feature in features_numerical:
    plt.figure(figsize=(15,10))
    sns.displot(data=data, x=feature, height=4, aspect=2, color='#800088')
    plt.show()
```

```python
data.drop(['Qualification'],axis=1 ,inplace=True)
```

```python
features_categorical = ['Gender','Nationality','Designation','Attendance','Site','Location','Motion Indication','Noise Detection','Infrared Sensor','Gas Sensor','Performance/KPI']
```

```python
for feature in features_categorical:
    plt.figure(figsize=(15,10))
    sns.displot(data=data, x=feature, height=4, aspect=2, color='#800088')
    plt.show()
```

```python

for feature in features_categorical:
    plt.figure(figsize=(15,10))
    sns.countplot(data=data, x=feature, palette='deep')
    plt.xticks(rotation=25)
    plt.show()
```

```python
for feature in features_numerical:
    plt.figure(figsize=(8,5))
    sns.barplot(data=data, x=feature, palette='deep')
    plt.yticks(rotation=18)
    plt.show()
```
